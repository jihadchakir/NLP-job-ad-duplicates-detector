```python
from gensim.models.doc2vec import Doc2Vec, TaggedDocument
import pandas as pd
import numpy as np
import nltk
from nltk.tokenize import word_tokenize
```

```python
data_org = pd.read_csv("new_data_jobs.csv")
data_org
```

```python
data_dv = pd.read_csv("final_jobs_data.csv")
data_dv
```

```python
print(data_dv['Title'][0])
print(data_org['Title'][0])
```

```python
def merge_strings(row):
    return ' '.join([str(val) for val in row if isinstance(val, str)])
```

```python
data_model = pd.DataFrame()
data_model['Merged_Column'] = data_dv.apply(merge_strings, axis=1)
data_model
```

```python
tagged_data=[TaggedDocument(words=word_tokenize(_d),tags=[data_dv['Title'][i]]) for i,_d in enumerate(data_model['Merged_Column'])]
```

```python
max_epochs=100
vec_size=100
alpha=0.025

model = Doc2Vec(vector_size=vec_size , alpha=alpha , min_alpha=0.00025 , min_count=1 , dm=1)

model.build_vocab(tagged_data)
```

```python
for epoch in range(max_epochs):
    print('iteration {0}'.format(epoch))
    model.train(tagged_data , total_examples=model.corpus_count , epochs=model.epochs)

model.save("d2v.model")
```

```python
print(model.infer_vector("developpeur full stack php casablanca".split()))
```

```python
similar_doc=model.docvecs.most_similar(data_dv['Title'][0])
print(similar_doc)
```

```python
Similar_jobs_list=[]
Similarity_value_list=[]
for i in range(len(data_dv['Title'])):
    similar_doc=model.docvecs.most_similar(data_dv['Title'][i])
    Similar_jobs_list.append(similar_doc[0][0])
    Similarity_value_list.append(similar_doc[1][1])

Similar_jobs=pd.DataFrame(Similar_jobs_list,columns=['Similar jobs'])
Similarity_Value=pd.DataFrame(Similarity_value_list,columns=['Similarity Value'])
```

```python
Doc2Vec_Sim_df=pd.concat([data_org['Title'],Similar_jobs,Similarity_Value],axis=1)
Doc2Vec_Sim_df
```

```python
jobs_list=[]
value_list=[]
for i in range(len(data_dv['Title'])):
    similar_doc=model.docvecs.(data_dv['Title'][i])
    jobs_list.append(similar_doc[0][0])
    value_list.append(similar_doc[1][1])

Similar_jobs=pd.DataFrame(jobs_list,columns=['Jobs'])
Similarity_Value=pd.DataFrame(value_list,columns=['Similarity Value'])
```
