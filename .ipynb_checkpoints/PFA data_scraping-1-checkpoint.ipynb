```python
from selenium import webdriver
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.common.by import By
from bs4 import BeautifulSoup
import requests
import numpy as np
import pandas as pd
```

```python
driver = webdriver.Chrome()
```

```python
links=[]
for i in range(1,33):
    url = 'https://www.rekrute.com/offres.html?p='+str(i)+'&s=3&o=1'
    driver.get(url)
    ele=driver.find_elements(By.CLASS_NAME, "titreJob")
    for e in ele:
        links.append(e.get_attribute("href"))
```

```python
Title=[]
Job_nature=[]
Experience=[]
Study=[]
Skills=[]
Contract=[]
Teletravail=[]
Date=[]
Employer=[]
Post=[]
Profile=[]
Adress=[]
```

```python
for link in links:
    driver.get(link)
    Title.append(driver.title)
    
    JN=driver.find_element(By.CLASS_NAME, "h2italic")
    Job_nature.append(JN.text)
    
    E=driver.find_element(By.XPATH,"//*[@title='Expérience requise']")
    Experience.append(E.text)
    
    St=driver.find_element(By.XPATH, '//li[@title="Niveau d\'étude et formation"]')
    Study.append(St.text)

    S=driver.find_elements(By.CLASS_NAME,"tagSkills")
    list=[]
    for e in S:
        list.append(e.text)
        resultat = '\n'.join(list)
    Skills.append(resultat)   
     
    C=driver.find_element(By.XPATH,"//*[@title='Type de contrat']")
    Contract.append(C.text)
    
    T=driver.find_element(By.XPATH, '//li[@title="Télétravail"]')
    Teletravail.append(T.text)
    
    D=driver.find_element(By.CLASS_NAME,"newjob")
    Date.append(D.text)
    
    html_content = driver.page_source
    soup1 = BeautifulSoup(html_content, 'html.parser')
    cards3 = soup1.find_all('div', class_='col-md-12 blc')
    
    if len(cards3)==9:
        e = cards3[1].text.strip()
        Employer.append(e)
    
        p = cards3[3].text.strip()
        Post.append(p)
    
        pr = cards3[4].text.strip()
        Profile.append(pr)
    else:
        e = cards3[1].text.strip()
        Employer.append(e)
    
        p = cards3[2].text.strip()
        Post.append(p)
    
        pr = cards3[3].text.strip()
        Profile.append(pr)
    
    try:
        A = driver.find_element(By.XPATH,"//*[@title='Région']")
        Adress.append(A.text)
    except:
        Adress.append(None)
```

```python
print(len(Title))
print(len(Job_nature))
print(len(Experience))
print(len(Study))
print(len(Skills))
print(len(Contract))
print(len(Teletravail))
print(len(Date))
print(len(Employer))
print(len(Post))
print(len(Profile))
print(len(Adress))
```

```python
data = np.array([Title,Job_nature,Experience,Study,Skills,Contract,Teletravail,Date,Employer,Post,Profile,Adress])
data = np.transpose(data)
df = pd.DataFrame(data, columns=['Title','Job_nature','Experience','Study','Skills','Contract','Teletravail','Date','Employer','Post','Profile','Adress'])
df
```

```python
df.to_csv('data.csv', index=False)
```
