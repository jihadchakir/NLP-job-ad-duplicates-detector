```python
import pandas as pd
import numpy as np
from selenium import webdriver
from bs4 import BeautifulSoup
from selenium.webdriver.chrome.options import Options
import requests
from selenium.webdriver.common.by import By
```

```python
title = []
locations = []
summary = []
experience_list = []
niveau_list = []
contrat_type = []
teletravail = []
delai_date = []
entreprises = []
postes = []
profiles = []
address_list = []
profile_list =[]
job_list = [] 
business_list = []
business = []
job = []
profile = []
address = []
personalite_temp = []
personalite = []
```

```python
driver = webdriver.Chrome()
```

```python
for pageno in range(1,27): 
    driver = webdriver.Chrome()
    driver.get("https://www.rekrute.com/offres-emploi-maroc.html") 
    
    
    
    url = 'https://www.rekrute.com/offres.html?p='+str(pageno)+'&s=3&o=1&languageId=6'
    driver.get(url) 
    print('scraping page ' + str(pageno))
    
    name = driver.find_elements(By.CLASS_NAME,'titreJob')    
    for s in range(len(name)):
        title.append(name[s].text)
    
    summaryItems = driver.find_elements(By.CLASS_NAME,'titreJob')
    job_links = [summaryItem.get_attribute("href") for summaryItem in summaryItems]
    


    
    for job_link in job_links:
        driver.get(job_link)
        
        close_button = driver.find_element(By.ID, "close-popup")
        close_button.click()
       
        job_description = driver.find_element(By.CLASS_NAME,"h2italic").text
        summary.append(job_description)

        experience_element = driver.find_element(By.XPATH, '//li[@title="Expérience requise"]')
        experience_list.append(experience_element.text)
        
        try:
            S=driver.find_elements(By.CLASS_NAME,"tagSkills")
            list=[]
            for e in S:
                list.append(e.text)
                resultat = ' - '.join(list)
            personalite.append(resultat)   
        except:
            personalite.append(None)
        
        try:
            region = driver.find_element(By.ID, "address")
            locations.append(region.text)
        except:
            locations.append(None)
            
            
        niveau = driver.find_element(By.XPATH, '//li[@title="Niveau d\'étude et formation"]')
        niveau_list.append(niveau.text)
        
        li_element_contract = driver.find_element(By.CSS_SELECTOR,'li[style="background: none; margin-right: 0px;"][title="Type de contrat"]')
        contract = li_element_contract.find_element(By.CSS_SELECTOR,'span.tagContrat').text.strip()
        contrat_type.append(contract)
            
        li_element_travail = driver.find_element(By.CSS_SELECTOR,'li[style="background: none; margin-right: 0px;"][title="Télétravail"]')
        travail = li_element_travail.find_element(By.CSS_SELECTOR,'span.tagContrat').text.strip()
        Ttravail = travail.replace("Télétravail : ","")
        teletravail.append(Ttravail)
        
        span_element_date = driver.find_element(By.CLASS_NAME, "newjob")
        date_element = span_element_date.find_element(By.TAG_NAME,"b")
        delai_date.append(date_element.text)
        
        entreprise_details = driver.find_element(By.XPATH, '//*[@id="recruiterDescription"]')
        entreprises.append(entreprise_details.text)

        html_content = driver.page_source
        soup1 = BeautifulSoup(html_content, 'html.parser')
        cards3 = soup1.find_all('div', class_='col-md-12 blc')

        if len(cards3) == 9:
            b = cards3[1].text.strip()
            res1 = b.replace("\n"," ").replace("\t"," ")
            business.append(res1)
            
            j = cards3[3].text.strip()
            res2 = j.replace("\n"," ").replace("\t"," ")
            job.append(res2)
            
            p = cards3[4].text.strip()
            res3 = p.replace("\n"," ").replace("\t"," ")
            profile.append(res3)
            
        else:
            
            b = cards3[1].text.strip()            
            res1 = b.replace("\n"," ").replace("\t"," ")
            business.append(res1)
            
            j = cards3[2].text.strip()
            res2 = j.replace("\n"," ").replace("\t"," ")
            job.append(res2)
            
            p = cards3[3].text.strip()
            res3 = p.replace("\n"," ").replace("\t"," ")
            profile.append(res3)

        
driver.close()
```

```python
data = np.array([title, summary, experience_list,locations, niveau_list, contrat_type, teletravail, delai_date, business, job, profile, personalite]) 
data = np.transpose(data) 
df = pd.DataFrame(data, columns=['Title', 'Description', 'Experience', 'Adress', 'School level', 'Contract Type', 'Teleworking', 'Deadline', 'Entreprise', 'Post information', 'Profile information', 'Personality'])
df
```
