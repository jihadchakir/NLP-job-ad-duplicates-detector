```python
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.feature_extraction.text import TfidfVectorizer
import pandas as pd
import numpy as np
import nltk
from nltk.tokenize import word_tokenize
import spacy
```

```python
nlp = spacy.load('fr_core_news_lg')
```

```python
data = pd.read_csv("final_jobs_data.csv")
```

```python
df_org=pd.read_csv('job_data.csv')
df_org
```

```python
data_transpose = data.transpose()
data_transpose
```

```python
def tokenize(data_in):
    tmp = []
    for j in data_in:
        doc = nlp(j)
        for token in doc:
            token = token.lemma_
            tmp.append(token)
    return tmp
```

```python
def calculate_jaccard(word_tokens1, word_tokens2):
    both_tokens = word_tokens1 + word_tokens2
    union = set(both_tokens)

    intersection = set()
    for w in word_tokens1:
        if w in word_tokens2:
            intersection.add(w)
    jaccard_score = len(intersection)/len(union)
    return jaccard_score
```

```python
tokens_jac=[]
for i in range(1231):
    token = tokenize(data_transpose[i])
    tokens_jac.append(token)
```

```python
text_elements = df_org['Title'].tolist()
jaccard_matrix = pd.DataFrame(index=text_elements, columns=text_elements)
```

```python
scores = []

for i in range(1231):
        for j in range(1231):
            score = calculate_jaccard(tokens_jac[i], tokens_jac[j])
            scores.append(score)
            jaccard_matrix.loc[df_org['Title'][i], df_org['Title'][j]] = score 
```

```python
jaccard_matrix = pd.DataFrame(jaccard_matrix.fillna(0))
jaccard_matrix
```

```python
jobs1=[]
jobs2=[]
for i in range(1231):
        for j in range(1231):
            jobs1.append(df_org['Title'][i])
            jobs2.append(df_org['Title'][j])
```

```python
jobs1_df=pd.DataFrame(jobs1,columns=['Job offer 1'])
jobs2_df=pd.DataFrame(jobs2,columns=['Job offer 2'])
scores_df = pd.DataFrame(scores, columns=['Score'])
```

```python
jaccard_sim_df=pd.concat([jobs1_df,jobs2_df,scores_df],axis=1)
```

```python
similar_jacc_df = jaccard_sim_df[(jaccard_sim_df['Score'] > 0.6)]
```
