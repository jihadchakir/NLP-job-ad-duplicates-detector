```python
import unicodedata
import string
import os as os
import pandas as pd
import numpy as np
import spacy
from spacy.lang.fr.stop_words import STOP_WORDS
import nltk
from nltk.stem.wordnet import WordNetLemmatizer
import spacy_transformers
from nltk.tokenize import word_tokenize
from nltk.corpus import wordnet
```

```python
jobs = pd.read_csv('job_data.csv')
```

```python
jobs_clean = pd.DataFrame()
```

```python
nlp = spacy.load('fr_core_news_lg')
```

```python
for i in range(1231):
    doc = nlp(jobs["Title"][i])

    for token in doc:
        if token.is_stop:
            print(token)
```

```python
def preprocess(text):
    doc = nlp(text)    
    no_stop_words = [token.text for token in doc if not token.is_stop]
    return " ".join(no_stop_words)
```

```python
for j in range(0,12):
    if j == 3 or j == 5 or j == 7 :
        continue
    jobs.iloc[:,j] = jobs.iloc[:,j].apply(preprocess) 
    print("colonne " + str(j))
```

```python
mising_values = jobs.isnull().sum()
mising_values
```

```python
for j in range(0,12):
    corpus = []
    corpus = jobs.iloc[:,j]

    if j==3:
        continue
    for i in range(len(corpus)) :
        tmp = str(corpus[i])
        tmp = tmp.lower()
        tmp = tmp.replace("’", "'")
        tmp = tmp.replace("\t"," ")
        tmp = tmp.replace("\n"," ")
        tmp = tmp.replace("/"," ")
        corpus[i] = tmp
         

    jobs_clean.loc[:, j] = corpus
```

```python
jobs_clean.rename(columns={0: "Title", 1: "Description", 2: "Experience", 3: "Address", 4: "School level",5: "Contract Type", 6: "Teleworking", 7: "Deadline", 8: "Entreprise", 9: "Post information", 10: "Profile information", 11: "Personality"}, inplace=True)
jobs_clean
```
